```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,100)
colors=["red", "green"]
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def tanh_derivative(x):
    return 1-tanh(x)**2
plt.plot(x, tanh(x), color=colors[0])
plt.plot(x, tanh_derivative(x), color=colors[1], linestyle="--")
plt.legend(["tanh","tanh_derivative"])
xlab=plt.xlabel("x")
ylab=plt.ylabel("y")
plt.show()
```
